## Data loading

### Subtask:
Load the advertising data from the provided CSV file into a pandas DataFrame.


**Reasoning**:
Import pandas and load the data into a DataFrame.



```python
import pandas as pd

df = pd.read_csv('advertising.csv')
df.head()
```

## Exploratory data analysis (eda)

### Subtask:
Perform EDA to understand the data, including checking for missing values, visualizing distributions, and exploring relationships between features.


**Reasoning**:
Check for missing values, generate descriptive statistics, and create histograms and scatter plots for the numerical columns to perform EDA.



```python
print("Missing values:")
print(df.isnull().sum())

print("\nDescriptive statistics:")
print(df.describe())

import matplotlib.pyplot as plt
import seaborn as sns

df.hist(figsize=(10, 8))
plt.tight_layout()
plt.show()

plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
sns.scatterplot(x='TV', y='Sales', data=df)
plt.title('TV vs Sales')

plt.subplot(1, 3, 2)
sns.scatterplot(x='Radio', y='Sales', data=df)
plt.title('Radio vs Sales')

plt.subplot(1, 3, 3)
sns.scatterplot(x='Newspaper', y='Sales', data=df)
plt.title('Newspaper vs Sales')

plt.tight_layout()
plt.show()
```

## Data preprocessing

### Subtask:
Prepare the data for machine learning by handling missing values (if any), encoding categorical variables (if any), and scaling numerical features.


**Reasoning**:
Prepare the data for machine learning by separating features and target, and then scaling the features.



```python
from sklearn.preprocessing import StandardScaler

# Separate features (X) and target (y)
X = df[['TV', 'Radio', 'Newspaper']]
y = df['Sales']

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Display the first few rows of the scaled features and the target
print("Scaled Features (X_scaled):")
print(X_scaled[:5])
print("\nTarget (y):")
print(y[:5])
```

## Model selection and training

### Subtask:
Choose a suitable machine learning model for regression (e.g., Linear Regression, Ridge, Lasso, or a tree-based model) and train it on the preprocessed data.


**Reasoning**:
Import the LinearRegression model, instantiate it, and fit it to the scaled feature data and the target variable.



```python
from sklearn.linear_model import LinearRegression

# Instantiate the LinearRegression model
model = LinearRegression()

# Fit the model to the scaled feature data and the target variable
model.fit(X_scaled, y)

print("Model training complete.")
```

## Model evaluation

### Subtask:
Evaluate the trained model's performance using appropriate metrics (e.g., Mean Squared Error, R-squared).


**Reasoning**:
Evaluate the trained model's performance using appropriate metrics.



```python
from sklearn.metrics import mean_squared_error, r2_score

# Make predictions on the scaled feature data
y_pred = model.predict(X_scaled)

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(y, y_pred)
print(f"Mean Squared Error (MSE): {mse}")

# Calculate R-squared score
r2 = r2_score(y, y_pred)
print(f"R-squared (R2) Score: {r2}")
```

## Prediction

### Subtask:
Use the trained model to make sales predictions.


**Reasoning**:
Use the trained model to make predictions on the scaled feature data and display the first few predictions.



```python
# Use the trained model to make predictions on the scaled feature data X_scaled
predictions = model.predict(X_scaled)

# Display the first few predicted values
print("First 5 predicted sales values:")
print(predictions[:5])
```

## Summary:

### Data Analysis Key Findings

*   There are no missing values in the dataset.
*   TV advertising appears to have a stronger linear relationship with Sales compared to Radio and Newspaper advertising based on scatter plot visualizations.
*   The trained Linear Regression model achieved a Mean Squared Error (MSE) of approximately 2.706.
*   The R-squared score of the model is approximately 0.903, indicating that the model explains about 90.3% of the variance in the sales data.

### Insights or Next Steps

*   The high R-squared value suggests that advertising expenditure on TV, Radio, and Newspaper are significant predictors of sales.
*   Further analysis could involve exploring non-linear relationships or interaction effects between the different advertising platforms to potentially improve the model's predictive power.
